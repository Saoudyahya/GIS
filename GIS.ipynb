```python
import geopandas as gpd
from shapely.geometry import Point, LineString
import folium
from geopy.distance import geodesic

# Replace with actual coordinates for Safi, Tangier, Genoa, and Belgium
safi_coordinates = (-9.2372, 32.2995)
tangier_coordinates = (-5.8000, 35.7667)
genoa_coordinates = (8.9463, 44.4056)
belgium_coordinates = (4.3517, 50.8503)

# Create a GeoDataFrame with points representing supply chain locations
data = {'Location': ['Safi', 'Tangier', 'Genoa', 'Belgium']}
latitude = [safi_coordinates[1], tangier_coordinates[1], genoa_coordinates[1], belgium_coordinates[1]]
longitude = [safi_coordinates[0], tangier_coordinates[0], genoa_coordinates[0], belgium_coordinates[0]]

# Create a GeoDataFrame
geometry_points = [Point(lon, lat) for lon, lat in zip(longitude, latitude)]
gdf_points = gpd.GeoDataFrame(data, geometry=geometry_points)

# Define the coordinate reference system (CRS) - WGS 84
gdf_points.crs = 'EPSG:4326'

# Create a folium map centered on Safi
m = folium.Map(location=[safi_coordinates[1], safi_coordinates[0]], zoom_start=4)

# Add points to the map
for idx, row in gdf_points.iterrows():
    folium.Marker([row.geometry.y, row.geometry.x], popup=row['Location']).add_to(m)

# Calculate great-circle paths between points
lines = []
for i in range(len(gdf_points) - 1):
    start_point = (gdf_points.iloc[i]['geometry'].y, gdf_points.iloc[i]['geometry'].x)
    end_point = (gdf_points.iloc[i + 1]['geometry'].y, gdf_points.iloc[i + 1]['geometry'].x)
    great_circle_line = geodesic(start_point, end_point).great_circle().points
    lines.append(great_circle_line)

# Add great-circle lines to the map
for line in lines:
    folium.PolyLine(locations=[(point[1], point[0]) for point in line], color='red').add_to(m)

# Display the map
display(m)
```
